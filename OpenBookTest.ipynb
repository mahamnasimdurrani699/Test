# Question no 2

```python
import pandas as pd

data = {
    'EmpID': [201,202,203,204,205,206,207,208,209,210],
    'Name': ['Ahmed','Maria','Saad','Areeba','Asim','Neha','Bilal','Zara','Hina','Umar'],
    'Department': ['HR','Finance','IT','HR','IT','Finance','HR','IT','Finance','IT'],
    'Attendance': [21,19,22,20,18,22,23,21,20,19],
    'Salary': [50000,60000,75000,52000,70000,62000,51000,72000,61000,69000]
}

df = pd.DataFrame(data)
```

```python
df
```

a: Group the data by Department and calculate the average attendance. Which department has the best attendance rate overall?

```python
group_data = df.groupby('Department')['Salary'].mean()
best_attendence=group_data.idxmax()
best_attendence
```

b: Filter employees with attendance less than 20 days and group them by
department.
How many employees in each department have poor attendance

```python
attendence_less = df[df['Attendance']<20].groupby('Department').size()
attendence_less
```

c: Group by Department and find both average and maximum salary.
Which department pays the highest salary and which one has the lowest
average salary?

```python
grouping = df.groupby('Department')['Salary'].agg(['max','min'])
high_salary = grouping['max'].idxmax()
low_salary = grouping['min'].idxmin()
high_salary,low_salary
```

d: Find out which employee(s) in the IT department have attendance above the
department’s average.

```python
attendance = df[df['Department'] == 'IT']['Attendance'].mean()

highest_attendance = df[(df['Department'] == 'IT') & (df['Attendance'] > attendance)]
highest_attendance

```

d: Create a report that shows total salary paid per department using groupby()
and sum(). Which department has the highest payroll burden?

```python
salary_paid = df.groupby('Department')['Salary'].sum()
salary_paid
```

e: Use groupby() and apply() to find employees in each department whose
attendance is above their departmental average.

```python
def above_avg(group):
    return group[group["Attendance"] > group["Attendance"].mean()]

above_avg_employees = df.groupby("Department").apply(above_avg)
above_avg_employees
```

# Question no 3

```python
raw_data = {
    'EmpID': [301, 302, 303, 304, 305, None, 307],
    'Name': ['Ahmed', None, 'Zara', 'Saad', 'Ahmed', 'Asim', 'Areeba'],
    'Department': ['HR', 'Finance', 'it', 'HR', 'HR', 'Finance', 'HR'],
    'JoinDate': ['2021-03-15', '2021-07-01', '2021-08-10', None, '2021-03-15', '2021/09/25', '2021-05-20'],
    'Salary': ['50000', None, '65000', '48000', '50000', '59000', 'Not Available']
}

df2 = pd.DataFrame(raw_data)
```

```python
df2
```

a: Identify and remove any duplicate records.

```python
remove = df2.drop_duplicates()
remove
```

b: Detects and handles missing values in Name, JoinDate, and Salary columns
appropriately.

```python
df2['Name'].fillna("Unknown", inplace=True)
df2['JoinDate'].fillna("1900-01-01", inplace=True)
df2['Salary'].replace("Not Available", None, inplace=True)
df2['Salary'] = pd.to_numeric(df2['Salary'], errors='coerce')
df2['Salary'].fillna(df2['Salary'].mean(), inplace=True)
```

d: Convert the JoinDate column to proper datetime format.

```python
df2['JoinDate'] = pd.to_datetime(df2['JoinDate'], errors='coerce')
df2['JoinDate']
```

d: Standardize the Department column values (e.g., correct lowercase to
uppercase).

```python
df2['Department'] = df2['Department'].str.upper()
df2['Department']
```

e: Drop any rows where EmpID is missing or not valid

```python
df2 = df2[df2['EmpID'].notnull()]
df2
```
